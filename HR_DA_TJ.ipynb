```python
import pandas as pd
data = pd.read_csv("Sheet.csv")
```

```python
manager=set(data["Account manager"])
client=len(set(data["Client Name"]))
```

```python
length=len(data)
visit={}
for i in xrange(0,length):
    year=int(data["Date of Contact"][i].split("-")[0])
    month=int(data["Date of Contact"][i].split("-")[1])
    client=data["Client Name"][i]
    
    if year in visit:
        if month in visit[year]:
            if client in visit[year][month]:
                visit[year][month][client]+=1
            else:
                visit[year][month][client]=1
        else:
            visit[year][month]={}
            visit[year][month][client]=1
    else:
        visit[year]={}
        visit[year][month]={}
        visit[year][month][client]=1
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
def plot_chart(dic):
    for key in dic:
        objects = range(1,13)
        y_pos = np.arange(len(objects))
        percentage = []
        for i in xrange(1,13):
            if i in dic[key]:
                client_month=len(dic[key][i])
            per_month=float(client_month)/client
            percentage.append(per_month)

        plt.bar(y_pos, percentage, align='center', alpha=0.5)
        plt.xticks(y_pos, objects)
        plt.ylabel('Percentage of Clients')
        plt.title(str(key))

        plt.show()
```

```python
plot_chart(visit)
```
